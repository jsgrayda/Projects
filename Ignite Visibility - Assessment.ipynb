```python
import numpy as np # need to import these packages for functionality
import pandas as pd

df = pd.read_csv('Excel Proficiency Test - Ignite Visibility.csv') #importing csv file

df.info() #checking for any missing values and the type of objects
```

# Task 1: Gen>Age

1. Segment data by campaign name, gender, age
2. Show these columns: Frequency, Spend, Link Click Through Rate, Cost per Sale, Conversion Rate

## Notes on Task 1

- I created a new variable called "Frequency as Num Type." First, I replaced the cells called "#DIV/0!" to zero (0) and converted the values into a numerical type (float) so that we can calculate the average for each column. 

- 'Conversion Rate' was a new column that I made by doing Purchases/Clicks

- Since the image of the pivot table gets cut off cus there are too many columns, one way to see the pivot table is to do: df.loc[''] and inside the brackets you can put the name of each campaign for example to identify info per campaign

```python
## create pivot table

df['Frequency as Num Type'] = df[' Frequency '].replace('#DIV/0!', 0).astype(float)

df['Conversion Rate'] = df['Purchases']/df['All Clicks']

pd.pivot_table(df, index=['Campaign Name','Gender','Age'], 
               values=['Frequency as Num Type','Spend', 'Link Click Through Rate(%)',
                      'Cost per Purchase','Conversion Rate'])
```
